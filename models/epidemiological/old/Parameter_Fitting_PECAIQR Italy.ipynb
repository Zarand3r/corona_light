```python
import numpy as np
from scipy.integrate import odeint
import matplotlib.pyplot as plt

# THIS IS A TEST OF ODEINT. USE AS REFERENCE OR IGNORE
def model(z,t):
    x = z[0]
    y = z[1]
    if t < 5.0:
        u = 0
    else:
        u = 2
    dxdt = (-x + u)/2.0
    dydt = (-y + x)/5.0
    dzdt = [dxdt,dydt]
    return dzdt

# initial condition
z0 = [0,0]

# number of time points
n = 401

# time points
t = np.linspace(0,40,n)
z = odeint(model,z0,t)
x = z[:,0]
y = z[:,1]

# plot results
plt.plot(t,x,'b-',label='x(t)')
plt.plot(t,y,'r--',label='y(t)')
plt.ylabel('values')
plt.xlabel('time')
plt.legend(loc='best')
plt.show()
```

```python
def peair(z, t, params):
    a_1 = params[0]
    a_2 = params[1]
    b_1 = params[2]
    b_2 = params[3]
    s_1 = params[4]
    s_2 = params[5]
    
    P = z[0]
    E = z[1]
    A = z[2]
    I = z[3]
    R = z[4]
    N = P + E + A + I + R
    
    dPdt = - ((a_1 * a_2) * E * P) / N
    dEdt = - ((b_1 * A + b_2 * I) * E) / N
    dAdt = (a_1 * E * P) / N + (b_1 * A * E) / N - (s_1 * A)
    dIdt = (a_2 * E * P) / N + (b_2 * I * E) / N - (s_2 * I)
    dRdt = s_1 * A + s_2 * I
    
    dzdt = [dPdt, dEdt, dAdt, dIdt, dRdt]
    return dzdt

n = 2000
params = [0.05, 0.05, 0.1, 0.1, 0.01, 0.01]
z0 = [750000, 250000, 10, 10, 0]
t = np.linspace(0,200,n)
z = odeint(peair,z0,t, args = (params, ))

P = z[:,0]
E = z[:,1]
A = z[:,2]
I = z[:,3]
R = z[:,4]


# plot results
plt.plot(t,P,'g--',label='P(t)')
plt.plot(t,E,'b--',label='E(t)')
plt.plot(t,A,'y--',label='A(t)')
plt.plot(t,I,'r:',label='I(t)')
plt.plot(t,R,'k--',label='R(t)')
plt.ylabel('# people')
plt.xlabel('time')
plt.legend(loc='best')
plt.show()
```
