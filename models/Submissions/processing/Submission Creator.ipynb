```python
import pandas as pd
import numpy as np
import os
```

```python
#Loading the default sample submission
Sample =  pd.read_csv('../../sample_submission.csv')
Sample['id'] = Sample['id'].astype(str)
```

```python
Sample.head()
```

```python
#Adding in relevant columns to help sort final submission
Sample['split'] = Sample['id'].str.rsplit('-',1)
Sample['Date'] = Sample['split'].str[0]
Sample['FIPS'] = Sample['split'].str[1]


Sample['Date'] = pd.to_datetime(Sample.Date)
Sample['Date1'] = Sample['Date'].dt.strftime('%m/%d/%Y')
```

```python
Sample.head()
```

```python
#############################################################
#Final Line: Writes out actual submission
Submission = Sample.set_index('id')
Submission.to_csv('Submission.csv', columns=['10','20','30','40','50','60','70','80','90'])
```
