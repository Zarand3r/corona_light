```python
import pandas as pd
import numpy as np
import os
```

```python
County_Beds = pd.read_csv('../../../data/us/hospitals/beds_by_county.csv', encoding='latin1')
County_ICU = pd.read_csv('../../../data/us/hospitals/icu_county.csv', encoding='latin1')
Cali_Data = pd.read_csv('../../../data/us/hospitals/ca_county_hospital_statistics.csv', encoding='latin1')

#State Bed Density (per 1000 people)
State_Pop = pd.read_csv('../../../data/us/demographics/state_populations.csv', encoding='latin1')
State_Bed = pd.read_csv('../../../data/us/hospitals/bed_densities.csv', encoding='latin1')
State_Bed.columns = State_Bed.columns.str.replace('bedspermille', 'bedsper1000')
```

```python
#Merging Hospital Data with ICU Data
County_Health = County_ICU.join(County_Beds.set_index('FIPS'), on='FIPS', lsuffix='_left', rsuffix='_right')
#Cleaning Merged Dataset
County_Health.columns = County_Health.columns.str.replace('icu_beds_left', 'icu_beds_Post')
County_Health.columns = County_Health.columns.str.replace('icu_beds_right', 'icu_beds_Pre')
County_Health = County_Health.drop(columns = ['Name'])
County_Health = County_Health[['FIPS','state','county','hospitals','staffed_beds','licensed_beds','icu_beds_Pre','icu_beds_Post']]

County_Health = County_Health.sort_values(by=['FIPS'])
County_Health.to_csv('County_Health.csv')
```

```python
#Cleaning State Bed Dataset
State_Bed = pd.merge(State_Bed, State_Pop, left_on='state', right_on='state')
State_Bed = State_Bed.drop(columns = ['country_x', 'country_y'])

State_Bed.insert(3, "Approx_Beds", State_Bed.bedsper1000*State_Bed.population/1000, True)
State_Bed = State_Bed.round(0)
State_Bed = State_Bed[['state','population','bedsper1000','Approx_Beds']]
State_Bed.to_csv('State_Bed.csv')
```

```python
#Cleaning California Data
Cali_Data.sort_values(by=['FIPS'])
Cali_Data = Cali_Data.drop(columns = ['Influenza_Season', 'State', 'CDPH_HAI_HCP_FLU_VACCINATION', 'Notes'])
Cali_Data.to_csv('Cali_Data.csv')
```
